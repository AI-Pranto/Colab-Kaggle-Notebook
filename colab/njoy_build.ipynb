## Clone and Install

```
! git clone --recurse-submodules https://github.com/njoy/NJOY21.git
% cd NJOY21/
! mkdir bin
% cd bin
! pip install cmake --upgrade
! cmake -DCMAKE_Fortran_COMPILER=`which gfortran` \
	-DCMAKE_CXX_COMPILER=`which clang++` \
	-DCMAKE_BUILD_TYPE=Debug \
	..
! make
! make install
```
